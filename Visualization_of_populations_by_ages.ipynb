```python

df = pd.read_csv('population_by_age_group.csv')
df.head()
```

```python

age_group_columns = ['65+', '25-64 years', '15-24 years', '5-14 years', '0-4 years']
age_group_totals = df[age_group_columns].sum()

plt.figure(figsize=(10, 6))
plt.bar(age_group_totals.index, age_group_totals.values, color='skyblue')
plt.title('Total Population by Age Group', fontsize=14)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.show()
```
